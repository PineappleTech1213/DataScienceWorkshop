```python
from math import floor
import pandas as pd
import datetime as dt
import numpy as np
```

```python

#read in datasets
patients = pd.read_csv('./datasets/patients.csv')
encounters = pd.read_csv('./datasets/encounters.csv')
patients.head()
```

```python
#change date to datetime
encounters['START'] = pd.to_datetime(encounters['START'])
encounters['STOP'] = pd.to_datetime(encounters['STOP'])
patients['DOB'] = pd.to_datetime(patients['BIRTHDATE'])
#filter encounters to be drug overdose after July 15, 1999
drug_overdose = encounters[(encounters['REASONDESCRIPTION'] == 'Drug overdose') & (encounters['START'] > '07-15-1999')]
#join patients and encounters by patient ids
do_patient_enct = encounters.merge(patients, left_on='PATIENT', right_on='Id')
#filter patients age to be 18 to 35 at time of encounters
do_patient_enct['AgeAtEncounter'] = do_patient_enct.apply(lambda x: floor((x['START'] - x['DOB']).days/365), axis=1)
do_pts_enct_filtered  = do_patient_enct[(do_patient_enct['AgeAtEncounter']>=18) & (do_patient_enct['AgeAtEncounter']<=35)]
print(do_pts_enct_filtered.head(10))
#rename some columns if needed.
#rename Id_x to EncounterId
#rename Id_y to PatientId
do_pts_enct_filtered.rename(columns={'Id_x':'ENCOUNTER_ID', 'Id_y':'PATIENT_ID','START':'HOSPITAL_ENCOUNTER_DATE',
                                     'AgeAtEncounter':'AGE_AT_VISIT'}, inplace=True)
```

```python
do_pts_enct_filtered.head(10)
```

```python
#part 2 create indicators
#cut the filtered table to needed columns
od_pts_enct = do_pts_enct_filtered[['ENCOUNTER_ID', 'PATIENT_ID', 'HOSPITAL_ENCOUNTER_DATE', 'STOP', 'AGE_AT_VISIT','DEATHDATE']]
#2.1 generate death indicators
od_pts_enct.head()
#change death date to date time 
od_pts_enct['DEATHDATE'] = pd.to_datetime(od_pts_enct['DEATHDATE'])
#generate death indicator
od_pts_enct['DEATH_AT_VISIT_IND'] = ''
for idx, row in od_pts_enct.iterrows():
    if row['DEATHDATE'] is pd.NaT:
        od_pts_enct.loc[idx, 'DEATH_AT_VISIT_IND'] = 'N/A'
    else:
        if row['DEATHDATE'] >= row['HOSPITAL_ENCOUNTER_DATE'] and row['DEATHDATE'] <=row['STOP']:
            od_pts_enct.loc[idx, 'DEATH_AT_VISIT_IND'] = '1'
        else:
            od_pts_enct.loc[idx, 'DEATH_AT_VISIT_IND'] = '0'

od_pts_enct.head(10)
```

```python
#2.2 get current med count
#read in medication tables
medications = pd.read_csv('./datasets/medications.csv')
medications.head()
def countPatientMedication(patient_id, startdate, medications):
    #count the number of medications the patient took before the encounter
    #check if the stop time is NA
    count = 0 
    pts_med = medications[medications['PATIENT'] == patient_id]  
    for index, row in pts_med.iterrows(): 
        if (row['STOP'] == 'NA' or row['STOP'] > startdate):
           count += 1
        else:
            continue
    print("the count is %d" % count)  
    return count

def getCurrentMedsCount(encounters, medications):
    #get a test dataset:
    # test = encounters.head(10)
    for index, row in encounters.iterrows():
        patient_id = row['PATIENT_ID']
        startdate = row['HOSPITAL_ENCOUNTER_DATE']
        encounters.loc[index, 'COUNT_CURRENT_MEDS'] =\
             countPatientMedication(patient_id,
             startdate, medications)
    print("finish counting current medications")
    return encounters


#change date to datetime
medications['START'] = pd.to_datetime(medications['START'])
medications['STOP'] = pd.to_datetime(medications['STOP'])
#find medications that started before July 15, 1999
part_med = medications[medications['PATIENT'].isin(do_pts_enct_filtered['PATIENT_ID'])][['START','STOP',
'PATIENT','ENCOUNTER','DESCRIPTION']]
print(part_med.info())
print(od_pts_enct.info())
# processPartialData(10,od_encounter, medications)
# print(x)
od_curr_med = getCurrentMedsCount(od_pts_enct, part_med)
```

```python
#2.3
#define check opiod functions
def checkActiveOpiod(drug_overdose, opiod_med):
    #check if the patient is using opiod at the time of drug overdose encounter
    # while n < 10 : do this... n+=1 to shorten the run time.
    drug_overdose['CURRENT_OPIOD_IND'] =\
                     drug_overdose.apply(lambda x: \
                        getOpiodUseStatus(x['PATIENT'],x['START'],opiod_med), axis=1)
    print("finish checking opiod use before overdose")
    return drug_overdose

def getOpiodUseStatus(patient,startdate,opiod_med):
    #check if the patient is using opiod at the time of drug overdose encounter
    for index2, row2 in opiod_med.iterrows():
        if row2['PATIENT'] == patient:
            if (row2['STOP'] == 'NA' or row2['STOP'] > startdate):
                print('find an opiod use')
                return 1
    return 0
opiod_list = ['Hydromorphone','Fentanyl',
    'Oxycodone-acetaminophen']
    #filter medications that contain opiod
part_med['CONTAIN_OPIOD'] = \
    part_med.apply(lambda x: containsAny(x['DESCRIPTION'], opiod_list), axis=1)
opiod_med = part_med[part_med['CONTAIN_OPIOD'] == True]
print(opiod_med.info())
od_opiod_active = checkActiveOpiod(od_curr_med, opiod_med)
```
